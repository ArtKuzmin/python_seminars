```python
from sympy import *
x = Symbol("x")
fx = x**3 - 50*x
gx = -x**4 + 88*x**2 - 241
```

1. Найти координаты точек пересечения
Приравниваем функции друг к другу и решаем, находя корни

```python
fgx = fx - gx
roots = solve(fgx) 
roots
```

```python
dct = {}
i = 1
for z in roots:
    dct[i] = {"X": round(N(z), 3), "y": round(N(-z**4 + 88*z**2 - 241), 3)}
    i+=1
dct
```
